```python

import json
from PIL import Image
from IPython.display import display
import random
```

```python
with open('/lustre/home/2001110054/Show-o/data/formalgeo7k/formalgeo7k_v2/custom_json/t2i/train.json', 'r') as f:
    train_data = json.load(f)
with open('/lustre/home/2001110054/Show-o/data/formalgeo7k/formalgeo7k_v2/custom_json/t2i/test.json', 'r') as f:
    test_data = json.load(f)
print(f'Num of train data: {len(train_data)}')
print(f'Num of test data: {len(test_data)}')
```

```python
total_idxs = [int(item['image'].split('/')[-1].replace('.png', '')) for item in train_data]
num_of_sample = 5
selected_idxs = random.sample(total_idxs, num_of_sample)
for idx in selected_idxs:
    image = Image.open(f'/lustre/home/2001110054/Show-o/data/formalgeo7k/formalgeo7k_v2/diagrams/{idx}.png').convert('RGB')
    selected_conversations = []
    for item in train_data:
        if item['image'].split('/')[-1].replace('.png', '') == str(idx):
            selected_conversations.append(item['conversations'])

    display(image.resize((384, 384)))
    for item in selected_conversations:
        for info in item:
            source = info['from']
            value = info['value']
            if value != '':
                print(f'{source}: {value}')
        
```
